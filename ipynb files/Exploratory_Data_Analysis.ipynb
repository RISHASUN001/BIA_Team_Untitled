```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import spacy

# Load the spaCy model for Named Entity Recognition
nlp = spacy.load("en_core_web_sm")  # Use 'en_core_web_sm' or another spaCy model

# Load the dataset
file_path = "news_cleaned.csv"  # Replace with the actual file path
data = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Information:")
data.info()
print("\nFirst few rows:")
print(data.head())

# 1. Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# 2. Data Cleaning
# Fill missing values in sentiment, subjectivity, and polarity with 0 if any
data['Sentiment'].fillna(0, inplace=True)
data['Subjectivity'].fillna(0, inplace=True)
data['Polarity'].fillna(0, inplace=True)

# 3. Prepare text data for word cloud
# Combine all the text data from your dataset (adjust the column name as needed)
text_data = " ".join(str(text) for text in data['Text'])  # Replace 'Text' with your actual column name

# Split the text into smaller chunks to avoid exceeding the character limit
chunk_size = 1000000  # Set the chunk size to the maximum length spaCy can handle
chunks = [text_data[i:i + chunk_size] for i in range(0, len(text_data), chunk_size)]

# Initialize an empty list to store entities
all_entities = []

# List of stopwords to manually filter out (you can extend this list as needed)
stopwords = {"said", "one", "two", "three", "someone", "something", "anyone", "anything", "first","second", "third", "year old","year", "old" }  # Example stopwords

# Process each chunk separately
for chunk in chunks:
    doc = nlp(chunk)  # Process the chunk with spaCy
    # Filter out stopwords from entities
    entities = [ent.text for ent in doc.ents if ent.text.lower() not in stopwords]
    all_entities.extend(entities)  # Add the entities from this chunk to the list

# Join the filtered entities into a single string for the word cloud
entities_text = " ".join(all_entities)

# Create and generate the word cloud using only the filtered entities
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(entities_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud of Extracted Entities')
plt.savefig("wordcloud_filtered.png")  # Save the plot
plt.show()
```

```python
# 6. Entity Analysis
# Extract and count unique entities from the Entities column
data['Entities'] = data['Entities'].apply(eval)  # Convert string to dictionary
entity_counts = {}
for entity_dict in data['Entities']:
    for key, values in entity_dict.items():
        for value in values:
            entity_counts[value] = entity_counts.get(value, 0) + 1
```

```python
# Sort and display top entities
sorted_entities = sorted(entity_counts.items(), key=lambda x: x[1], reverse=True)[:10]
entity_names, entity_values = zip(*sorted_entities)

plt.figure(figsize=(10, 6))
sns.barplot(x=list(entity_values), y=list(entity_names), palette='viridis')
plt.title('Top 10 Entities')
plt.xlabel('Frequency')
plt.ylabel('Entity')
plt.savefig("top_entities.png")  # Save the plot
plt.show()
```

```python
# 7. Analysis by Link
# Check how many articles are from each source
plt.figure(figsize=(14, 10))  # Increase width for longer links
data['Link'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title('Top Sources by Article Count')
plt.xlabel('Count')
plt.ylabel('Source')
plt.xticks(rotation=0)  # Ensure x-axis labels are readable
plt.tight_layout()  # Adjust layout to fit labels properly
plt.savefig("analysis_link.png", bbox_inches="tight")  # Save with proper bounding
plt.show()
```

```python
# Save Correlation Matrix
correlation_matrix = data[['Polarity', 'Subjectivity']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig("correlation_matrix.png")  # Save the plot
plt.show()
```
